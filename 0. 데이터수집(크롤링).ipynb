```python
import requests
import lxml.html
from selenium.webdriver import Chrome
import re
from urllib.parse import urljoin
import time
from selenium.webdriver import Chrome
from selenium import webdriver
```

```python
url = 'https://www1.president.go.kr/petitions/?c=0&only=2&page={}&order=2'
```

```python
from fake_useragent import UserAgent
ua = UserAgent()
```

```python
options = webdriver.ChromeOptions()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36")
```

```python
browser = Chrome(options=options)
```

```python
titles= []
counts = []
infos = []
contents = []
```

```python


user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36"

headers = {'User-Agent' : user_agent}

for page in range(4214,10000):
    try:
        browser.get(url.format(page))
        time.sleep(0.3)
        petitions = browser.find_elements_by_css_selector('div.bl_body a.cb.relpy_w')[8:15]
        for petition in petitions:
            links = []
            links.append(petition.get_attribute('href'))
            for link in links:
                user_agent = ua.random
                headers = {'User-Agent' : user_agent}
                res = requests.get(link,headers = headers)
                root = lxml.html.fromstring(res.text)
                titles.append(root.cssselect('h3.petitionsView_title'))
                print(root.cssselect('h3.petitionsView_title'))
                counts.append(root.cssselect('.petitionsView_count'))
                infos.append(root.cssselect('.petitionsView_info_list li'))
                contents.append(root.cssselect('div.View_write'))
                time.sleep(1)
        print(page)

    except Exception as ex: #  
        print(page,'   ', ex)
        break

    
    
```

```python
a = []
for i in titles:
    for x in i:
        a.append(x.text_content())
       
        
```

```python
b = []
for i in contents:
    if len(i)>=2:
        b.append(re.sub(r"\s+", " ",i[1].text_content()))
    else:
        for x in i:
            b.append(re.sub(r"\s+", " ",x.text_content()))
       
        
```

```python
c = []
for i in counts:
    for x in i:
        c.append(''.join(re.findall(r'\d',x.text_content())))
    
        
```

```python
category = []
startdate= []
enddate = []
person = []
for i in infos:
    for x in i:
        data = x.text_content() 
        if re.findall(r'',data):
            category.append(re.sub(r'','',x.text_content()))
        elif re.findall(r'',data):
            startdate.append(x.text_content())
        elif re.findall(r'',data):
            enddate.append(x.text_content())
        elif re.findall(r'',data):
            person.append(x.text_content())
```

```python
import pandas as pd
df = pd.DataFrame({
    '' : a,
    '' : b,
    '' : c,
    '' : category,
    '' : startdate,
    '' : enddate,
    '' : person
})
```

```python
df.to_csv('data.csv',encoding='utf8',index=False)
```
